```python
import os
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter

# Directory containing PDF files
pdf_directory = '/content/'

# Initialize an empty list to store all documents
all_docs = []

# Initialize the text splitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

# Iterate through all files in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):  # Check if the file is a PDF
        file_path = os.path.join(pdf_directory, filename)

        # Load the PDF file
        pdf_loader = PyPDFLoader(file_path)
        pdf_pages = pdf_loader.load_and_split()

        # Split the PDF pages into chunks
        docs = text_splitter.split_documents(pdf_pages)

        # Add the chunks to the list of all documents
        all_docs.extend(docs)

# Print the total number of chunks created
print(f"Total number of chunks created: {len(all_docs)}")

# Optionally, print the first chunk for verification
print("First chunk:")
print(all_docs[0])
```

```python
db=FAISS.from_documents(all_docs,embedding)
```

```python
db
```

```python
retriever=db.as_retriever()
from langchain.agents.agent_toolkits import create_retriever_tool
tool=create_retriever_tool(
    retriever,
    "Competitive RAG",
    "searches and returns related writing dataregarding the query"

)
tools=[tool]
```

```python
!pip install langchain # Make sure langchain is installed

from langchain.agents.agent_toolkits.conversational_retrieval.openai_functions import create_conversational_retrieval_agent # Import the necessary function

# ... your existing code ...

agent_executor = create_conversational_retrieval_agent(llm, tools, verbose=True) # Now this line should work
```

```python
input="Write a 200 word essay on sufi movement in india"
output=agent_executor.invoke(input=input)
print(output)
```

```python
parameters = {
    "temperature": 0.2,  # Creativity (lower = more deterministic)
    "max_length": 400,   # Limit the length of generated text
    "top_p": 0.9,        # Controls diversity (top-p sampling)
    "num_return_sequences": 1  # Number of outputs you want to generate
}
```

```python
from IPython.display import display, Markdown
system_prompt = "You are an educational assistant designed to help students improve their writing skills for competitive exams. Follow these guidelines:\n"
"1. Answer questions in a style as given in  the uploaded PDFs.\n"
"2. Use simple, clear , concise language.\n"
"3. If the answer isn't in the documents, retrieve a well strucutred answer .'\n"
"4. Do not speculate, assume, or invent information.\n"
"5. Maintain a professional tone and organize responses clearly (e.g., bullet points, step-by-step explanations).\n"
"6. Encourage follow-up questions by asking if further clarification is needed.\n"
"7. Provide examples to clarify concepts when helpful.\n"
"8. Keep answers concise, focused, and exam-friendly."
improv = "Provide a precise and well-structured answer based on the context above. Ensure your response is easy to understand, includes examples where necessary, and is formatted in a way that students can use it for exams. If applicable, ask if the student needs further clarification."
input = f"{system_prompt}  {improv} Write an essay in 250 words  religious minorities in india and constitution, structure it in 3-4 paragraphs"
output = agent_executor.invoke(input=input,**parameters)

# Extract the AI's response from the chat history
ai_response = output['chat_history'][-1].content

# Display the AI's response as Markdown
display(Markdown(ai_response))
```
